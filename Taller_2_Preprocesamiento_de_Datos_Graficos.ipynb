7. Columnas redundantes

```python
# Columnas redundantes identificadas con la varianza

from sklearn.feature_selection import VarianceThreshold
import pandas as pd

# Selección de columnas numéricas
columnas_interes = df.select_dtypes(include=['int64']).columns
X = df[columnas_interes]

# Aplicar VarianceThreshold para elminiar datos con vairanza igual a 0
selector = VarianceThreshold()  
X_reduced = selector.fit_transform(X)

# Obtener las columnas que se mantienen
selected_columns = columnas_interes[selector.get_support()]

print(f"Dimensión original: {X.shape}")
print(f"Dimensión después de VarianceThreshold: {X_reduced.shape}")
print(f"Columnas eliminadas: (varianza cero): {set(columnas_interes) - set(selected_columns)}")
print(f"Columnas conservadas: {selected_columns.tolist()}")
```

10. identificación de valores atipicos con gráfica y funciones

```python
# Identificación de Valores Atípicos
import matplotlib.pyplot as plt

# diagrama de cajas y bigotes para identifición de valores atípicos
plt.boxplot(df['edad'])
plt.title('Distribución de Edad')
plt.ylabel('Edad')
plt.show()
```

```python
# Técnica de media y Varianza pra eliminación de valores atípicos

limite_superior= df['edad'].mean () + 3 * df['edad'].std ()
limite_inferior = df['edad'].mean () - 3 * df['edad'].std ()
df_sin_atipicos = df[(df['edad'] < limite_superior) & (df['edad'] > limite_inferior)]
df_sin_atipicos
```

```python
# Vlaores mim y max y limites

import numpy as np
print("valor Mínimo =", np.min(df['edad']))
print("Valor Máximo =", np.max(df['edad']))
print("Limite inferior = ", limite_inferior)
print("Limite Superior = ", limite_superior)
```

```python
# Técnica de los 1 y 99 percentiles para eliminación de valores atípicos

sup = df['edad'].quantile(.99)
inf = df['edad'].quantile(.01)
datos_sin_atipicos = df[(df['edad'] < sup) & (df['edad'] > inf)]
datos_sin_atipicos

```
